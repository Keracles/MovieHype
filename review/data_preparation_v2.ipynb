```python
def modif_date(row):
    date = row['date']
    liste = date.split('-')
    liste[0] = '2000'
    string = ''
    for i in liste:
        string += i + '-'
    row['date'] = string[:-1]
    return row
```

```python
df = df.apply(modif_date, axis = 1)
```

```python
def add_season(row):
    dico_saison = {'printemps': 0, 'été': 1, 'automne' : 2, 'hiver' : 3}
    if row['date'] >= pd.to_datetime('2020-12-21') or row['date'] <= pd.to_datetime('2020-03-19'):
        row['saison'] = dico_saison['hiver']
    if pd.to_datetime('2020-03-20') <= row['date'] <= pd.to_datetime('2020-06-20'):
        row['saison'] = dico_saison['printemps']
    if pd.to_datetime('2020-06-21') <= row['date'] <= pd.to_datetime('2020-09-22'):
        row['saison'] = dico_saison['été']
    else : 
        row['saison'] = dico_saison['automne']
    return row
    
```

```python
df = df.apply(add_season, axis = 1)
```

```python
list_keys = ['realisateur', 'casting', 'compagnies_production']

def str_to_list(df, key):
    df_new = df

    for k in key: #on parcourt les clés à modifier spécifiées en argument de la fonction

        for ind in range(df_new.shape[0]): #on parcourt toutes les lignes du Dataframe

            string = df_new[k].iloc[ind] #on récupère la valeur à modifier pour chaque ligne

            if type(string) == str : # on vérifie si c'est bien une string
                
                #création de la nouvelle variable de type liste qui remplacera l'ancienne

                liste_finale = []

                #remplissage de la variable de type liste en faisant des opérations sur la string

                list_string = string[1:-1].split(',')
                for i,j in enumerate(list_string):
                    if i == 0:
                        mot = j[1:-1]
                        if mot not in liste_finale:
                            liste_finale.append(mot)
                    else:
                        mot = j[2:-1]
                        if mot not in liste_finale:
                            liste_finale.append(mot)
                
                #affectation de la nouvelle valeur de type liste à l'élément modifié du Dataframe

                df_new.at[ind,k] = liste_finale

    return df_new
```

```python
df = str_to_list(df,list_keys)
```

```python
from difflib import SequenceMatcher
```

```python
def comparaison_string(str1, str2): #Renvoie la ressemblance en deux strings
    S = SequenceMatcher(None,  str1, str2)
    if S.ratio() > 0.8 :
        return True
    else :
        return False
```

```python
def calcul_scores_casting(row):
    liste_score = []
    for i in list(row['casting']):
        score = 0
        for j in list(df_actor['Actor_Name']):
            if comparaison_string(i, j):
                score = int(df_actor[df_actor['Actor_Name'] == j]['rating'])
                break
            else :
                pass
        liste_score.append(score)
    liste_score_sorted = sorted(liste_score)       
    sum_score = sum(liste_score_sorted)
    centre = len(liste_score_sorted) // 2
    mediane_score = (liste_score_sorted[centre] + liste_score_sorted[~centre]) / 2   
    row['casting_score_sum'] = sum_score
    row['casting_score_mean'] = sum_score/len(liste_score)
    row['casting_score_med'] = int(mediane_score)
    row['casting_score_min'] = liste_score_sorted[0]
    row['casting_score_max'] = liste_score_sorted[len(liste_score_sorted)-1]
    return row

df_test = df.apply(calcul_scores_casting, axis = 1)
df_test.head()
```

```python
df_test.to_csv('../data/dataset_new_score.csv')
```
