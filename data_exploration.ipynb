### fonction permettant de transformer les données en string dans le DataFrame en liste de string pour faciliter leurs manipulations

```python
list_keys = ['genres', 'production_countries', 'realisateur', 'casting', 'compagnies_production', 'langues']

def str_to_list(df, key):
    df_new = df

    for k in key: #on parcourt les clés à modifier spécifiées en argument de la fonction

        for ind in range(df_new.shape[0]): #on parcourt toutes les lignes du Dataframe

            string = df_new[k].iloc[ind] #on récupère la valeur à modifier pour chaque ligne

            if type(string) == str : # on vérifie si c'est bien une string
                
                #création de la nouvelle variable de type liste qui remplacera l'ancienne

                liste_finale = []

                #remplissage de la variable de type liste en faisant des opérations sur la string

                list_string = string[1:-1].split(',')
                for i,j in enumerate(list_string):
                    if i == 0:
                        mot = j[1:-1]
                        if mot not in liste_finale:
                            liste_finale.append(mot)
                    else:
                        mot = j[2:-1]
                        if mot not in liste_finale:
                            liste_finale.append(mot)
                
                #affectation de la nouvelle valeur de type liste à l'élément modifié du Dataframe

                df_new.at[ind,k] = liste_finale

    return df_new
```

### Fonction pour mettre en forme le synopsis pour faciliter son étude

```python
def suppr_caractere_string(string):

    chars = ['.', ',', '!','?', ':', ';', "'", '"', '(', ')', '[', ']', '{', '}', '-'] 
 
    new_string = string.translate(str.maketrans('', '', ''.join(chars)))

    return new_string
```

```python
def suppr_elem_list(liste):
    
    list_elem_to_suppr = ['', 'the', 'be','to', 'of', 'and', 'a', 'in', 'that', 'have', 'I', 'i', 'it', 'for', 'not', 'on', 'with', 'he', 'as', 'you', 'do', 'at', 'this', 'but', 'his', 'by', 'from','they', 'we', 'say', 'her', 'she', 'or', 'an', 'will', 'my', 'one', 'all', 'would', 'there', 'their', 'what','so', 'up', 'out','if', 'about', 'who', 'get', 'which', 'go', 'me', 'when', 'make', 'can', 'like', 'time', 'no', 'just', 'him', 'know', 'take', 'people', 'into', 'year', 'your', 'good', 'some', 'could', 'them', 'see', 'other', 'than', 'then', 'now', 'look', 'only', 'come', 'its', 'over', 'think', 'also', 'back', 'after', 'use', 'two', 'how', 'our', 'work', 'first', 'well', 'way :', 'even', 'new', 'want', 'because', 'any', 'these', 'give', 'day', 'most', 'us', 'agree', 'allow', 'answer', 'ask', 'be', 'become', 'begin', 'believe', 'borrow', 'bring', 'build', 'buy', 'call', 'carry', 'change', 'choose', 'close', 'come', 'cut', 'decide', 'do', 'dream', 'drink', 'drive', 'eat', 'explain', 'fall', 'feel', 'fill', 'find', 'finish', 'follow', 'fly', 'forget', 'get']

    new_list = liste

    for elem in list_elem_to_suppr:

        new_list = list(filter((elem).__ne__, new_list))

    return new_list
```

```python
def mise_en_forme_synopsis(df):
    
    df_new = df

    for i in range(df_new.shape[0]):

        synopsis_brut = df_new.synopsis.iloc[i]

        if type(synopsis_brut) == str:

            syn_sans_ponct = suppr_caractere_string(synopsis_brut)

            syn_sans_maj = syn_sans_ponct.lower()

            synopsis_list = syn_sans_maj.split(' ')

            new_synopsis_list = suppr_elem_list(synopsis_list)

            df_new.at[i, 'synopsis'] = new_synopsis_list
    
    return df_new
```

### Fonction permettant de compter le nombre de film par genre en renvoyant un dicitonnaire

```python
def nb_film_genre(df):

    dico_genre = {}

    for i in range(df.shape[0]):

        list_genre = df.iloc[i].genres

        for genre in list_genre:

            if genre in list(dico_genre.keys()):
                dico_genre[genre] += 1

            else:
                dico_genre[genre] = 1
                
    return dico_genre
```

### Fonction permettant de compter le nombre de film en fonction du mois de sortie

```python
def nb_film_mois(df):

    dico_mois = {} #créaction d'un dictionnaire pour stoker le nombre de film par mois

    df['date'] = pd.to_datetime(df['date']) #modification du format de la date de string à datetime64[ns]

    for i in range(df.shape[0]): #parcourt de toutes les lignes du DataFrame

        mois_de_sortie = df['date'].dt.month_name().iloc[i] #récupération du mois de sortie à partir de la date

        if type(mois_de_sortie) == str : #vérification que le mois de sorti est bien une string pour ne pas avoir de problème avec une clés nan
             
            # comptage du mois de sorti dans le dictionnaire 
            if mois_de_sortie in list(dico_mois.keys()):
                dico_mois[mois_de_sortie] += 1
            else:
                dico_mois[mois_de_sortie] = 1

    return dico_mois
```

### Fonction permettant de compter le nombre de film en fonction de la période de sortie

```python
def nb_film_period(df):

    new_df = df

    dico_period = {} 

    new_df['date'] = pd.to_datetime(new_df['date']) 

    new_df['year'] = df.date.dt.year 

    new_df['periode'] = np.digitize(new_df['year'], [1970, 1990, 2000, 2010, 2015], right=True)

    period_labels = ['1900-1970', '1970-1990','1990-2000', '2000-2010', '2010-2015', '2015-2022']
    
    for i in range(new_df.shape[0]):

        period = new_df.periode.iloc[i]

        period_label = period_labels[period]

        if period_label in list(dico_period.keys()):
            dico_period[period_label] += 1
        else:
            dico_period[period_label] = 1
    

    return dico_period

```

### fonction permettant d'obtenir le nombre de films par tranche de bénefice

```python
def nb_film_benef(df):

    new_df = df

    dico_benef = {}
    dico_rentable = {'rentable' : 0, 'non rentable' : 0}

    new_df['benef'] = new_df['recette'] - new_df['budget']

    new_df['tranche_benef'] = np.digitize(new_df['benef'], [-10000000,-1000000,0,50000, 100000, 500000,1000000, 50000000, 100000000, 1000000000], right=True)

    benef_labels = ["< - 10000000","[-10000000,-1000000]", "[-1000000,0]","[0,50000]", "[50000, 100000]", "[100000, 500000]", "[500000,1000000]", "[1000000, 50000000]", "[50000000, 100000000]","[100000000, 1000000000]", "> 1000000000"]
    
    for i in range(new_df.shape[0]):

        benef = new_df.tranche_benef.iloc[i]

        if new_df.recette.iloc[i] - 2*new_df.budget.iloc[i] < 0 :
            dico_rentable['non rentable'] += 1
        else :
            dico_rentable['rentable'] += 1

        benef_label = benef_labels[benef]

        if benef_label in list(dico_benef.keys()):
            dico_benef[benef_label] += 1
        else:
            dico_benef[benef_label] = 1
    

    return dico_benef, dico_rentable
```

### Fonction permettant d'afficher la recette moyenne des films par mois de sorti

```python
def recette_mois(df):
    
    recette_mois = {}

    df['date'] = pd.to_datetime(df['date']) #modification du format de la date de string à datetime64[ns]
    
    for i in range(df.shape[0]): #parcourt de toutes les lignes du DataFrame

        mois_de_sortie = df['date'].dt.month_name().iloc[i] #récupération du mois de sortie à partir de la date

        if type(mois_de_sortie) == str : #vérification que le mois de sorti est bien une string pour ne pas avoir de problème avec une clés nan
             
            # ajout de la recette du film en fonction du mois de sortie dans le dictionnaire 
            if mois_de_sortie in list(recette_mois.keys()):
                recette_mois[mois_de_sortie] += df['recette'].iloc[i]
            else:
                recette_mois[mois_de_sortie] = df['recette'].iloc[i]
        
    nb_film = nb_film_mois(df) #récupération du nombre de film par mois
        
        #pour chaque mois on divise la somme des recettes par le nombre de films
    for mois in list(recette_mois.keys()):
        recette_mois[mois] = recette_mois[mois]/nb_film[mois]
        
    return recette_mois
```

### recette par pays par années

```python
def recette_pays(df):
  list_dico_recette_pays = []
  for j in range(2000,2023):
    dico_recette_pays = {} #créaction d'un dictionnaire pour stoker le nombre de film par pays de production

    for i in range(df.shape[0]): #parcourt de toutes les lignes du DataFrame

      pays_prod_list = df['production_countries'].iloc[i] #récupération des pays de production

      for pays in pays_prod_list : #on parcourt les pays de production d'un film

        # comptage du pays de production dans le dictionnaire
        if (pays in list(dico_recette_pays.keys()) and df['date'].iloc[i].year==j) :
          dico_recette_pays[pays] += df['recette'].iloc[i]
        if (pays not in list(dico_recette_pays.keys()) and df['date'].iloc[i].year==j) :
          dico_recette_pays[pays] = df['recette'].iloc[i]

    list_dico_recette_pays.append(dico_recette_pays)

  return list_dico_recette_pays
```

### Fonction permettant d'afficher un dictionnaire sous forme de bar chart

```python
def dict_to_bar(dico):

    if type(dico) == dict:
        dict_to_display = {}

        for k, v in sorted(dico.items(), key=lambda x: x[1]):
            dict_to_display[k] = v

        labels = list(dict_to_display.keys())
        x = list(dict_to_display.values())
        nb_keys = len(labels)
        plt.figure(figsize=(nb_keys*2,nb_keys))
        plt.bar(labels, x)
        plt.show()

    elif type(dico)== list:

        plt.figure(figsize=(50,100))
        for i in range(len(dico)):

            dict_to_display = {}

            for k, v in sorted(dico[i].items(), key=lambda x: x[1]):
                dict_to_display[k] = v
            
            labels = list(dict_to_display.keys())
            x = list(dict_to_display.values())
            plt.subplot(len(dico),1,i+1)
            plt.bar(labels, x)
        
    else :
        print("l'argument n'est ni une liste ni un dictionnaire")
```

### fonction permettant de faire des box plot des variables numériques

```python
def boxplot(df):
    minimum1 = df['recette'].min()
    maximum1 = df['recette'].max()

    minimum2 = df['budget'].min()
    maximum2 = df['budget'].max()

    minimum3 = df['vote_count'].min()
    maximum3 = df['vote_count'].max()

    minimum4 = df['suite'].min()
    maximum4 = df['suite'].max()

    minimum5 = df['duree'].min()
    maximum5 = df['duree'].max()

    minimum6 = df['vote_average'].min()
    maximum6 = df['vote_average'].max()

    boxplot = df

    boxplot['recette'] = ((df['recette'] - minimum1) / (maximum1 -minimum1)) 
    boxplot['budget'] = ((df['budget'] - minimum2) / (maximum2 -minimum2))
    boxplot['vote_count'] = ((df['vote_count'] - minimum3) / (maximum3 -minimum3))
    boxplot['suite'] = ((df['suite'] - minimum4) / (maximum4-minimum4))
    boxplot['duree'] = ((df['duree'] - minimum5) / (maximum5-minimum5))
    boxplot['vote_average'] = ((df['vote_average'] - minimum6) / (maximum6-minimum6))

    plt.figure(figsize = (20,20))
    plt.title('boxplot movies')
    plt.xticks(rotation = 60)
    plt.subplot(2,3,1)
    sns.boxplot(data=boxplot['budget'],fliersize=5)
    plt.subplot(2,3,2)
    sns.boxplot(data=boxplot['suite'],fliersize=5)
    plt.subplot(2,3,3)
    sns.boxplot(data=boxplot['vote_average'],fliersize=5)
    plt.subplot(2,3,4)
    sns.boxplot(data=boxplot['vote_count'],fliersize=5)
    plt.subplot(2,3,5)
    sns.boxplot(data=boxplot['duree'],fliersize=5)
    plt.subplot(2,3,6)
    sns.boxplot(data=boxplot['recette'],fliersize=5)
```
